## Question 2:

```python
from selenium.common.exceptions import NoSuchElementException
import time

brand = []
product = []
delivery = []
price = []
URL = []

for i in range(3):
    
    name_tag = driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none']/h2/a/span")
    prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
    del_tag = driver.find_elements_by_xpath('//div[@class="a-row s-align-children-center"]/span[2]/span[2]')
    urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    
    for i in urls:
        URL.append(i.get_attribute('href'))
        
    for i in name_tag:
        if i.text.split()[0] == None:
            brand.append('-')
        else:
            brand.append(i.text.split()[0])

        if i.text.split(' (')[0] == None:
            product.append('-')
        else:
            product.append(i.text.split(' (')[0])

    for i in prices:
        price.append(i.text)
        
    for i in del_tag:
        if i.text == None:
            delivery.append('-')
        else:
            delivery.append(i.text)

    next_link = driver.find_element_by_xpath("//ul[@class = 'a-pagination']/li[@class='a-last']/a")
    driver.get(next_link.get_attribute('href'))
    time.sleep(2)
```

```python
search = driver.find_element_by_id("twotabsearchtextbox")
search.clear()
search.send_keys(item)
```

```python
Ratings = []
total_rate = []

for j in ['https://www.amazon.in/s?k=earphones&page=2&qid=1614249941&ref=sr_pg_1', 'https://www.amazon.in/s?k=earphones&page=3&qid=1614261100&ref=sr_pg_2', 'https://www.amazon.in/s?k=earphones&page=4&qid=1614263385&ref=sr_pg_3']:
    for url in URL:
        driver.get(url)
        try:                  
            rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
            rate.click()                                                      
            rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
            Ratings.append(rating.text)
            rates = driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary']")
            total_rate.append(rates.text.strip())
        except NoSuchElementException   as e:
            Ratings.append("NO rating")
            total_rate.append('0 count')
            
    driver.get(j)
    time.sleep(2)
```

```python
amz = pd.DataFrame({'Brand':brand[:60], 'Product Name':product[:60], 'Price':price[:60], 'Delivery Date':delivery[:60], 
                    'Rating':Ratings[:60], 'No.of rating':total_rate[:60], 'Product URL':URL[:60]})
```

## Question 4:

```python
driver.get('https://www.flipkart.com/')
```

```python
search_job = driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search_job.send_keys('smart phones')
```

```python
search_btn = driver.find_element_by_xpath("//div[@class='col-12-12 _2oO9oE']/button")
search_btn.click()
```

```python
brand = []
phone = []
colour = []
price = []
ram = []
rom = []
front_camera = []
rear_camera = []
processor = []
core_processor = []
display_size = []
display_res = []
battery = []
prod_url = []

name_tag = driver.find_elements_by_xpath("//div[@class='_4rR01T']")
ram_tag = driver.find_elements_by_xpath('//div[@class="fMghEO"]/ul[1]/li[1]')
display_tag = driver.find_elements_by_xpath('//div[@class="fMghEO"]/ul[1]/li[2]')
camera_tag = driver.find_elements_by_xpath('//div[@class="fMghEO"]/ul[1]/li[3]')
battery_tag = driver.find_elements_by_xpath('//div[@class="fMghEO"]/ul[1]/li[4]')
proc_tag = driver.find_elements_by_xpath('//div[@class="fMghEO"]/ul[1]/li[5]')
price_tag = driver.find_elements_by_xpath('//div[@class="_30jeq3 _1_WHN1"]')
url_tags = driver.find_elements_by_xpath("//a[@class = '_1fQZEK']")

for i in name_tag:
    if i.text.split()[0] == None:
        brand.append('-')
    else:
        brand.append(i.text.split()[0])
    
    if i.text.split(' (')[0] == None:
        phone.append('-')
    else:
        phone.append(i.text.split(' (')[0])
        
    if i.text.split('(')[1].split(',')[0] == None:
        colour.append('-')
    else:
        colour.append(i.text.split('(')[1].split(',')[0])
        
    if i.text.split(',')[1][:-1] == None:
        rom.append('-')
    else:
        rom.append(i.text.split(', ')[1][:-1])

for i in ram_tag:
    if i.text.split(' | ')[0] == None:
        ram.append('-')
    else:
        ram.append(i.text.split(' | ')[0])
        
for i in display_tag:
    if i.text == None:
        display_res.append('-')
        display_size.append('-')
    else:
        display_size.append(' '.join(i.text.split()[:4]))
        display_res.append(' '.join(i.text.split()[4:]))

for i in camera_tag:
    if i.text == None:
        front_camera.append('-')
        rear_camera.append('-')
    else:
        rear_camera.append(i.text.split(' | ')[0])
        if 'Front' in i.text.split(' | ')[1]:
            front_camera.append(i.text.split(' | ')[1].split()[0])
        else:
            front_camera.append(i.text.split(' | ')[1])        
        
for i in battery_tag:
    if i.text == None:
        battery.append('-')
    else:
        battery.append(' '.join(i.text.split()[:2]))
        
for i in proc_tag:
    if 'core' not in i.text.lower():
        core_processor.append('-')
    else:
        core_processor.append(i.text.split('core')[0])
        
    if 'processor' not in i.text.lower():
        processor.append('-')
    else:
        processor.append(i.text)
        
for i in price_tag:
    if i.text == None:
        price.append('-')
    else:
        price.append(i.text)

for i in url_tags:
    prod_url.append(i.get_attribute('href'))
```

```python
smartphones = pd.DataFrame({'Brand':brand, 'Product Name':phone, 'Price':price, 'Colour':colour, 'Ram':ram, 'Rom':rom, 
                            'Front Camera':front_camera, 'Rear Camera':rear_camera, 'Battery': battery, 
                            'Display Size': display_size, 'Display Resolution': display_res, 
                            'Processor': processor, 'Core Processor': core_processor, 'Product URL':prod_url})
```

```python
smartphones
```

```python
smartphones.to_csv('Smartphone_details.csv', index=False, encoding='utf-8')
```

## Question 6:

```python
driver.get('https://trak.in/')
```

```python
driver.find_element_by_xpath("//nav[@class = 'main-menu-container bs-pretty-tabs bs-pretty-tabs-initialized']/ul[1]/li[9]/a").get_attribute('href')
```

```python
date = []
startup = []
industry = []
sub_vert = []
place = []
investor = []
investment_type = []
amt = []

for t in ['tablepress-50', 'tablepress-49', 'tablepress-48']:
    row = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr')
    
    for i in range(len(row) +1):

        # to traverse through the table row 
        date_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[2]')
        startup_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[3]/a')
        industry_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[4]')
        sub_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[5]')
        place_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[6]')
        investor_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[7]')
        investment_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[8]')
        amt_tag = driver.find_elements_by_xpath(f'//table[@id="{t}"]/tbody/tr["{i}"]/td[9]')

    for i in date_tag:
        date.append(i.text)

    for i in startup_tag:
        startup.append(i.text)

    for i in industry_tag:
        industry.append(i.text)

    for i in sub_tag:
        sub_vert.append(i.text)

    for i in place_tag:
        place.append(i.text)

    for i in investor_tag:
        investor.append(i.text)

    for i in investment_tag:
        investment_type.append(i.text)

    for i in amt_tag:
        amt.append(i.text)
```

```python
funding = pd.DataFrame({'Date':date, 'Start-up Name':startup, 'Industry':industry, 'Sub-Vertical':sub_vert, 
                        'Place':place, 'Investor Name':investor, 'Investment Type':investment_type, 'Amount': amt})
```
