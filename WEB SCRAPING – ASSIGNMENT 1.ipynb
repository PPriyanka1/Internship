### 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
# To get the content object from th Webpage
page2 = requests.get('https://www.imdb.com/list/ls091520106/')
```

```python
#### Parse the source code by Beautiful Soup with html parser as the source code is in html format

soup2 = BeautifulSoup(page2.content, 'html.parser')
```

```python
# use prettify function to see the source code in readable format

print(soup2.prettify())
```

```python
# We have created 3 empty lists for 3 features which we have to extract

name_list, rating_list, =  [], []
```

```python
# Extract the header and stored in hlist
hlist = soup2.find_all('h3', class_='lister-item-header')
```

```python
# Fetching the Movie name from the 'a' tag
for i in hlist:
    for j in i.find_all('a'):
        name_list.append(j.get_text())
```

```python
# Retriving the year of release value from html tagss
yor = soup2.find_all('span', class_="lister-item-year text-muted unbold")
```

```python
# adding the data to the list from beautiful_soup objects
yor_list = []
for i in yor:
    if 'I' in i.get_text():
        yor_list.append(i.get_text().replace('(I) ', '').lstrip('(').rstrip(')')) # One of the value having '(I) (2017) -> converted this value'
        continue
    yor_list.append(i.get_text().lstrip('(').rstrip(')'))
```

```python
# Converting string to int
yor_list = [int(i) for i in yor_list]
```

```python
# Fetching the div tag from webpage
div_list = soup2.find_all('div', class_='ipl-rating-star small')
```

```python
# Fetching the rating from the div_list
for i in div_list:
    for j in i.find_all('span', class_='ipl-rating-star__rating'):
        rating_list.append(j.get_text())
```

```python
# Converting the string to float
rating_list = [float(i) for i in rating_list]
```

```python
import pandas as pd

imdb= pd.DataFrame({})
imdb['Movie Name'] = name_list[:100]
imdb['IMDB Rating'] = rating_list[:100]
imdb['Year of Release'] = yor_list[:100]
```

```python
imdb
```

```python
imdb.to_csv('IMDB_Top_Rated_Movies.csv')
```

### 3. Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page3 = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
```

```python
#### Parse the source code by Beautiful Soup with html parser as the source code is in html format

soup3 = BeautifulSoup(page3.content, 'html.parser')
```

```python
# Extract the body and stored in body_list

body_list = soup3.find_all('tbody', class_='lister-list')
```

```python
# Extract movie_name
movie_list_india = []
rating_list_india = []
yor_list_india = []

for i in body_list:
    # Fetching the movie_names from the webpage and store the movie names to the list 
    for movie in i.find_all('a'):
        movie_list_india.append(movie.get_text().strip())
    
    # Fetching the year f release from the webpage and store to the list
    for yr in i.find_all('span', class_='secondaryInfo'):
        yor_list_india.append(yr.get_text().lstrip('(').rstrip(')'))
    
    # Fetching the rating from the webpage and store to the list
    for rate in i.find_all('strong'):
        rating_list_india.append(rate.get_text())
```

```python
# removing the extra empty string from the list
movie_list_india = [st for st in movie_list_india if st != '']
```

```python
#Converting string to int
yor_list_india = [int(i) for i in yor_list_india]
```

```python
import pandas as pd
imdb_india = pd.DataFrame({})

imdb_india['Indian_Movie_Name'] = movie_list_india[:100]
imdb_india['Indian_Released_Year'] = yor_list_india[:100]
imdb_india['Indian_Rating'] = rating_list_india[:100]
```

```python
imdb_india
```

```python
# save the DataFrame to the csv
imdb_india.to_csv('IMDB_Top_Rated_Indian_Movies.csv')
```

```python
import pandas as pd
men_top_team = pd.DataFrame({})
men_top_team['POS'] = men_team_rank[:10]
men_top_team['Team'] = men_team_name[:10]
men_top_team['Matches'] = men_team_matches[:10]
men_top_team['Points'] = men_team_points[:10]
men_top_team['Rating'] = men_team_rating[:10]
```

```python
period = []
short_desc = []
desc = []
temp = []
for i in weather_soup_list:
    for day in i.find_all('p', class_='period-name'):
        period.append(day.get_text().strip())
    
    for short in i.find_all('p', class_='short-desc'):
        short_desc.append(short.get_text())
    
    for summ in i.find_all('img'):
        if 'alt' in summ.attrs:
            desc.append(str(summ.attrs['alt']))
    
    for tp in i.find_all('p', class_='temp'):
        temp.append(tp.get_text())
```

```python
import pandas as pd
weather = pd.DataFrame({})

weather['Period'] = period[:9]
weather['Temperature'] = temp[:9]
weather['Short Description'] = short_desc[:9]
weather['Description'] = desc[:9]
```
