```python
driver.get('https://www.bcci.tv/')
```

```python
try:
    url3 = driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']/div[1]/ul/li[1]/a")
    driver.get(url3.get_attribute('href'))
    
except ElementNotVisibleException as e:
    print('Exception Raised: ',e)
except UnexpectedTagNameException as e:
    print('Tag Name not found:', e)
```

```python
match = []
series = []
place = []
date = []
month = []
time = []

try:
    match_tag = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__format']")
    series_tag = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
    place_tag = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span")
    date_tag = driver.find_elements_by_xpath("//span[@class='fixture__date']")
    month_tag = driver.find_elements_by_xpath("//span[@class='fixture__month']")
    time_tag = driver.find_elements_by_xpath("//span[@class='fixture__time']")


    for i in match_tag:
        match.append(i.text)

    for i in series_tag:
        series.append(i.text)

    for i in place_tag:
        place.append(i.text)

    for i in date_tag:
        date.append(i.text)

    for i in month_tag:
        month.append(i.text)

    for i in time_tag:
        time.append(i.text)

    result = ', '.join('% s  % s' % i for i in zip(date, month))
    date_list = result.split(', ')
    

except UnexpectedTagNameException as e:
    print('Tag Name not found:', e)
```

```python
cricket = pd.DataFrame({'Match':match, 'Series': series, 'Place': place, 'Date':date_list, 'Time': time})
cricket
```

```python
driver.get('https://archive.ics.uci.edu/ml/index.php')
```

```python
try:
    link = driver.find_element_by_xpath("//table[@bgcolor='#003366']/tbody/tr/td[2]/span[2]/a")
    driver.get(link.get_attribute('href'))

except ElementNotVisibleException as e:
    print('Execption Raised: ', e)
except UnexpectedTagNameException as e:
    print('Tag Name not found:', e)
```

```python
dataset_name = []
datatype = [] 
task = []
attribute_type = []
instances = []
attribute = []
year = []

try:
    dataset_name_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
    datatype_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]/p")
    task_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]/p")
    attribute_type_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]/p")
    instances_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]/p")
    attribute_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]/p")
    year_tag = driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]/p")

    for i in dataset_name_tag:
        if i.text == ' ':
            dataset_name.append('NA')
        else:
            dataset_name.append(i.text.strip())

    for i in datatype_tag:
        if i.text == ' ':
            datatype.append('NA')
        else:
            datatype.append(i.text.strip())

    for i in task_tag:
        if i.text == ' ':
            task.append('NA')
        else:
            task.append(i.text.strip())

    for i in attribute_type_tag:
        if i.text == ' ':
            attribute_type.append('NA')
        else:
            attribute_type.append(i.text.strip())

    for i in instances_tag:
        if i.text == ' ':
            instances.append('NA')
        else:
            instances.append(i.text.strip())

    for i in attribute_tag:
        if attribute == ' ':
            attribute.append('NA')
        else:
            attribute.append(i.text.strip())
        
    for i in year_tag:
        if i.text == ' ':
            year.append('NA')
        else:
            year.append(i.text.strip())
        
except UnexpectedTagNameException as e:
    print('Tag Name not found:', e)
```

```python
uci = pd.DataFrame({'DataSet Name': dataset_name, 'DataType': datatype[1:], 'Task':task[1:], 'Attribute Type':attribute_type[1:], 'Instance Type':instances[1:], 'Attribute':attribute[1:], 'Year':year[1:]})
uci
```
