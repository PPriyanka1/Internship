### Removing the Outliers

```python
# Check the Outliers
from scipy.stats import zscore

# Take without categorical
z_score = np.abs(zscore(mcd))
print(mcd.shape)

mcd_final = mcd[(z_score < 3).all(axis=1)]
print(mcd_final.shape)
```

### Model Training

Now our data are ready!

According to the data, our Target variable 'label' is discrete, so our model is Classification.

Let's now begin to train out Classification model!

We will need to first split up our data into an X array that contains the features to train on, and y array with the target variable

```python
# Separationg independent & dependent variables

df_x = mcd_final.drop(columns=['label'])
df_y = mcd_final['label']
```

```python
# Scaling the input variable

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x = sc.fit_transform(df_x)
```

#### Build the Model

```python
#Split data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, df_y, test_size = 0.3, random_state = 42)
```

```python
# Importing the model library

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier

# Import Error Metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
model = [LogisticRegression(), KNeighborsClassifier(), SVC(), GaussianNB(), DecisionTreeClassifier(), RandomForestClassifier(), AdaBoostClassifier(), GradientBoostingClassifier(), BaggingClassifier()]

acc = []
for m in model:
    print('\n*************** %s Model ***************\n' % m)
    m.fit(x_train, y_train)
    print('Score of %s Model is %s' % (m, m.score(x_train, y_train)))
    predm = m.predict(x_test)
    acc_sc = round(accuracy_score(y_test, predm)*100, 2)
    print('Accuracy_score: %s' % acc_sc)
    print('Confusion Matrix : \n', confusion_matrix(y_test, predm))
    print('\nClassification Report : \n', classification_report(y_test, predm))
    print('F1 Score: ',round(f1_score(y_test, predm, average='micro')*100,2))
    
    acc.append(acc_sc)
```

```python
#Let's compare the accuracies of each model!

models = pd.DataFrame({
    'Model': ['LogisticRegression', 'KNeighborsClassifier', 'SVC', 'GaussianNB', 'DecisionTreeClassifier', 'RandomForestClassifier', 'AdaBoostClassifier', 'GradientBoostingClassifier', 'BaggingClassifier']
,
    'Score': acc})

models.sort_values(by='Score', ascending=False)
```

Observation:

From above models, **RandomForestClassifier, GradientBoostingClassifier, BaggingClassifier** gives the best results

```python
mean = []
mini = []
std = []
maxi = []
for m in model:
    print('\n*************** %s Model ***************\n' % m)
    cv_score = cross_val_score(m, x, df_y, cv=5, scoring = 'accuracy')
    print('Score of %s Model is %s' % (m, cv_score))
    print('Mean score: ', cv_score.mean())
    print('Std score : ', cv_score.std())
    print('Min score : ', cv_score.min())
    print('Max score : ', cv_score.max())
    mean.append(cv_score.mean())
    mini.append(cv_score.min())
    std.append(cv_score.std())
    maxi.append(cv_score.max())
```

```python
#Let's compare the accuracies of each model!

cv = pd.DataFrame({
    'Model': ['LogisticRegression', 'KNeighborsClassifier', 'SVC', 'GaussianNB', 'DecisionTreeClassifier', 'RandomForestClassifier', 'AdaBoostClassifier', 'GradientBoostingClassifier', 'BaggingClassifier']
, 'Mean': mean, 'Std':std, 'Min':mini, 'Max': maxi})

cv
```

We use the **GridSearchCV** to find the best parameters and check the score and after that we finalized our model

### Using GridSearchCV to find best Parameters

```python
params = { 
    'n_estimators': [10, 100, 200],
    'max_depth':[3,4,5,6],
    'max_features': ['auto', 'sqrt']
}

rf = RandomForestClassifier()

clf = GridSearchCV(rf, params)
clf.fit(x, df_y)
print('Best Parameters: ', clf.best_params_)
```

```python
# GraidentBoostingClassifier
gbc = GradientBoostingClassifier()

clf = GridSearchCV(gbc, params)
clf.fit(x, df_y)
print('Best Parameters: ', clf.best_params_)
```

```python
# BaggingClassifier
params = { 
    'n_estimators': [10, 100, 200]}

bc = BaggingClassifier()

clf = GridSearchCV(bc, params)
clf.fit(x, df_y)
print('Best Parameters: ', clf.best_params_)
```

### Applying results to the model

```python
RF = RandomForestClassifier(max_depth= 6, max_features= 'sqrt', n_estimators = 10)
GBC = GradientBoostingClassifier(criterion='mse', max_depth= 6, max_features= 'sqrt', n_estimators = 200)
BC = BaggingClassifier(n_estimators=200)
```

```python
models = []
models.append(('RandomForestClassifier', RF))
models.append(('GradientBoostingClassifier', GBC))
models.append(('BaggingClassifier', BC))
```

```python
Model = []
score = []
CVS = []
ROC_Score = []

for name, model in models:
    print('*********** %s **********\n' %(name))
    Model.append(name)
    
    model.fit(x_train, y_train)
    print(model)
    pred = model.predict(x_test)
    print('\n')
    
    # Accuracy Score
    AS = accuracy_score(y_test, pred)
    print('Accuracy Score: ', AS)
    score.append(AS*100)
    print('\n')
    
    # Cross Val Score
    cross_val = cross_val_score(model, x, df_y, cv=10, scoring='accuracy').mean()
    print('Cross Val Score :', cross_val)
    CVS.append(cross_val*100)
    print('\n')
    
    # ROC_AUC_Score
    from sklearn.metrics import roc_curve, roc_auc_score, auc
    fpr, tpr, threshold = roc_curve(y_test, pred)
    roc_auc = auc(fpr, tpr)
    print('ROC_AUC Score :', roc_auc)
    ROC_Score.append(roc_auc*100)
    print('\n')
    
    # Confusion Matric
    
    print('Confusion matrix: ', confusion_matrix(y_test, pred))
    
    # Classification Report
    print('\nClassification Report : \n', classification_report(y_test, pred))
    
    # Confusion Matrix Heatmap
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt='.0%', cmap='cool'))
    plt.subplot(912)
    plt.title(name)
    
    # ROC_Curve

    plt.plot(fpr, tpr, label='AUC=%.2f'%roc_auc)
    plt.plot([0, 1], [0,1], 'r--')
    plt.plot(fpr, tpr, label='GaussianNB')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='Lower Right')
    plt.show()
```

```python
# storing the result in DataFrame
result = pd.DataFrame({'Model':Model, 'Accuracy_Score':score, 'Cross_Val_Score':CVS, "ROC_AUC_Score":ROC_Score})
result
```
